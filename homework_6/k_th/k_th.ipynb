# Quick select

SOlution:

1. Выберается опорный элемент (pivot) из массива.
2. Массив делиться на три части:
- leftArr: Элементы больше опорного элемента.
- midArr: Элементы равные опорному элементу.
- rightArr: Элементы меньше опорного элемента.
3. Проверка условия:
- Если k меньше или равен размеру leftArr, то ищите k-й элемент в leftArr.
- Если k больше размера leftArr + размера midArr, то скорректируйте k и ищите в rightArr.
- В противном случае верните опорный элемент (это и есть k-й по величине элемент).


````python
def quick_select(arr: List[int | float], k: int) -> int | float:
    pivot = arr[0]

    leftArr = [x for x in arr if x > pivot]

    midArr = [x for x in arr if x == pivot]

    rightArr = [x for x in arr if x < pivot]

    if k <= len(leftArr):
        return quick_select(leftArr, k)
    if len(leftArr) + len(midArr) < k:
        return quick_select(rightArr, k - len(leftArr) - len(midArr))

    return pivot
````

````python
random_list = [int(x) for x in np.random.randint(low=-100, high=100, size=7)]
k = 3

print("og list:", random_list, "\nsorted list:", list(enumerate(sorted(random_list, reverse=True), start=1)))

print(f"max {k}-th element", quick_select(random_list, k=k))
````
