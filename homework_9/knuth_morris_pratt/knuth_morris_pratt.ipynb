# Solution

Логика:
1) Вычисляем префикс-функцию для паттерна (в дальнейшем последовательность рі). Объявляем две переменные для хранения индексов позиции в строке и подстроке соответственно (в дальнейшем і и j). j - значение индекса первого элемента. 
2) Проверяем совпадение символов на і позиции в строке и на j позиции в подстроке.
-  Символы совпадают -> пункт 3
-  Символы не совпадают -> пункт 4
3) Если j равен индексу последнего символа подстроки -> нашли подстроку. В противном случае увеличиваем і и j на единицу -> пункт 2
4) Выполняем проверку индексов і и j.
- Индекс і равен или больше индекса последнего символа строки -> не нашли паттерн.
- Индекс і равен индексу первого символа в подстроке, то увеличить значение і на единицу -> пункт 2.
- Индекс j не равен индексу первого символа в подстроке. В таком случае установить значение j = pi[j-1] -> пункт 2.

Т.к. вычисление префикс-функции для паттерна происходит за O(m), а в алгоритме мы проходимся по всем символам в строке, то 

**Time Complexity:** O(n + m), где 
- n - кол-во символов в строке
- m - кол-во символов в паттерне

```python
def prefix_function(text):
    n = len(text)
    pi = [0] * n
    for i in range(1, n):
        j = pi[i - 1]
        while j > 0 and text[j] != text[i]:
            j = pi[j - 1]
        if text[i] == text[j]:
            j += 1
        pi[i] = j
    return pi


def kmp_search(text, sub_text):
    j = 0
    pi = prefix_function(sub_text)
    for i in range(len(text)):
        while j > 0 and text[i] != sub_text[j]:
            j = pi[j - 1]
        if text[i] == sub_text[j]:
            j += 1
        if j >= len(sub_text):
            return i - j + 1
    return None
```

```python
text = "AAATGAACGAAAATCTGT"
sub_text = "ACGA"
kmp_search(text, sub_text)
```

```python
txt = "aabaacaadaabaaba"
pat = "aaba"
kmp_search(txt, pat)
```
